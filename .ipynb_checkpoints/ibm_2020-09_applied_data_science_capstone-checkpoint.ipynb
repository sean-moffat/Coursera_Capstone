<h2>1. Introduction</h2>
  
First responders (EMT's, fire fighters and police) across the United States have challenging careers where there is additional risk not needed when traveling to and from an incident calls. In 2017, collisions involving fire department emergency vehicles accounted for 1,005 injuries to firefighters in the U.S. according to [Evarts, Ben. (2018, November 1). U.S. Firefighter Injuries in 2017. NFPA Journal.](https://www.nfpa.org/News-and-Research/Publications-and-Media/NFPA-Journal/2018/November-December-2018/Features/US-Firefighter-Injuries-2017). Some police traffic accidents are a direct result of first responders disregarding their department's policies or disregarding a direct order to terminate a pursuit. In Chicago, police pursuits have increased in frequency and resulted in more damage. 

As a result of fatal policy pursuit crash on June 3, 2020 [(Charles, Sam. (2020, June 4). Lightfoot to change police pursuit policy after fatal crash. Chicago Sun-Times.)](https://chicago.suntimes.com/city-hall/2020/6/4/21280780/chicago-police-department-pursuit-policy-fatal-crash), Chicago's mayor as made swift changes to the police pursuit policy; this [policy was changed in August 2020](http://directives.chicagopolice.org/directives/data/a7a57be2-1291920c-54712-9192-aecb02ef17e0c47d.pdf?hl=true)

First responders involved in traffic accidents en route to incident calls have immeasurable costs to those requesting help along with those answering the call for help. It is in everyone's interest to know and understand the facts of this situation. With a shared understanding of the facts, the general public, elected officials and first responders can develop data-driven policies to mitigate accidents during first responders' incident calls.

This analysis will use **supervised machine learning (SML)** to predict the severity of accidents. In addition, through non-SML statistical modeling I will attempt to determine if first responders / all other vehicles have a statistically significant impact on accident severity. 


<h2>2. Data</h2>

I used one data source for this analysis, [The City of Chicago's open data portal](https://data.cityofchicago.org/). Specifically, this analysis used data sets from Traffic Crashes subject area under the Transportation genre. There are three seperate traffic crashes data sets: *Crashes*, *Vehicles* and *People*. The traffic crashes data sets report information on each traffic crash on city streets within the City of Chicago limits and under the jurisdiction of Chicago Police Department (CPD). 

Each of the Traffic Crashes data set file contain data from March 3, 2013 - September 7, 2020, although complete citywide data coverage didn't begin until September 1, 2017. These data sets were joined using the CRASH_RECORD_ID attribute. 

Lastly, a crash represents a collision between a car and one or more other cars, pedestrian, bicyclist, or fixed object. Below is a brief description of each data set, followed by Tables 1 and 2 which report traffic accidents figures by vehicle use and time. All data sets were accessed on September 8, 2020 at 11:45 a.m.

I collected *Crashes* data from The City of Chicago's open data portal for [Traffic Crashes - Crashes](https://data.cityofchicago.org/Transportation/Traffic-Crashes-Crashes/85ca-t3if) data. This file contains crash data where each record in the file represents a unique traffic crash. *__2.B. Vehicles__* 

I collected *Vehicles* data from The City of Chicago's open data portal for [Traffic Crashes - Vehicles](https://data.cityofchicago.org/Transportation/Traffic-Crashes-Vehicles/68nd-jvt3) data. This file contains information about vehicles involved in traffic crashes. Each record in the file represents a vehicle. There is a many-to-one relationship between the Vehicles and Crashes data sets because it is possible that a single traffic accident involves more than one vehicle.  

To accociate a vehicle use to the severity of traffic accident involved merging Crashes and Vehicles data sets. I mapped the vehicle use to the traffic accident data by slicing the vehicles data set into four new dataframes, each frame containing unique (based on CRASH_RECORD_ID) traffic accidents involving: ambulance, fire, police and all other vehicles; Table 2 reports high level figures (in aggregate across time) on traffic accidents with injuries and with fatalities, by vehicle use. 

The 'VEHICLE_USE' attribute will be used to identify ambulance, fire, police (representing first responders) and all other vehicles.  

I am interested in the relationship between severity of traffic accidents and involvement of first responder vehicles. To analysis this, I engineered a FLAG_FIRST_RESPONDER categorical feature (1 = first responder and 0 = all other vehicles). There are two motivating factors driving this interest as discussed in the Introduction.

```python
# Data wrangling . . . , re: convert to 'date' attributes from source file date time object
df['CRASH_DATE_V02'] = pd.to_datetime(df['CRASH_DATE'])

# Using date time attribures created above, create new date only attributes 
df['CRASH_DATE-DATE_ONLY'] = df['CRASH_DATE_V02'].dt.date

# Create a Year Month data element . . . will produce YYYY-MM
df['CRASH_DATE_YEAR_MONTH'] = df['CRASH_DATE_V02'].dt.strftime('%Y-%m')
# Create a Year data element . . . will produce YYYY-MM
df['CRASH_DATE_YEAR'] = df['CRASH_DATE_V02'].dt.strftime('%Y')
#df['YEAR_OF_CRASH'] = df['CRASH_DATE_V02'].map(lambda x: x.strftime('%Y')) # ALTERNATE METHOD 
# Create a Month data element . . . will produce MM
df['CRASH_DATE_MONTH'] = df['CRASH_DATE_V02'].dt.strftime('%m')
# Create a Day data element . . . will produce DD
df['CRASH_DATE_DAY'] = df['CRASH_DATE_V02'].dt.strftime('%d')


grouped_df = df.groupby('CRASH_DATE_YEAR').size()
print('Distribution of CRASH_DATE_YEAR across all records in the original data set')
print(grouped_df)
print("----- OUTPUT BREAK -----")
print('')
print('Oldest date in Crashes data set')
print(df.CRASH_DATE_V02.min())
print('')
print('Most recent date in Crashes data set')
print(df.CRASH_DATE_V02.max())
```

<h5><i>Missing data</i></h5>
<p>
The original data set contains many features, however, not all features will be useful for predicting accident severity. At some point, I will need to drop features that don’t make sense to include in models predicting accident severity. Before I drop features, I want to understand how many records are missing data points from the current feature set. 
</p>

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum())/df.isnull().count().sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total','Percent'], sort=False).sort_values('Total', ascending=False)
missing_data.head(40)
```

Running a missing data analysis indicates that 11 of the 49 original features are missing data for at least 50% of the records. With so many missing records, it would be inappropriate to attempt to fill in missing values by __mean, frequency or other functions__. 

Decision: Drop records where at least 50% of the attributes have missing values

```python
df.drop(['WORKERS_PRESENT_I',
         'DOORING_I', 
         'WORK_ZONE_TYPE', 
         'WORK_ZONE_I', 
         'PHOTOS_TAKEN_I', 
         'STATEMENTS_TAKEN_I', 
         'NOT_RIGHT_OF_WAY_I', 
         'CRASH_DATE_EST_I', 
         'INTERSECTION_RELATED_I', 
         'HIT_AND_RUN_I', 
         'LANE_CNT' 
        ], axis=1, inplace=True)

print("What's the shape of our dataframe after dropping records where at least 50% of the attributes have missing values")
print('Number of features: {}'.format(df.shape[1]))
print('Number of examples: {}'.format(df.shape[0]))
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum())/df.isnull().count().sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total','Percent'], sort=False).sort_values('Total', ascending=False)
missing_data.head(15)
```

We have a number of missing records in the "injury" attributes. Our goal is to predict predict the severity of an accident. We will drop these rows because any data entry without injury data cannot be used for prediction; therefore any row now without injury data is not useful to us

```python
# simply drop whole row with NaN in "injury" columns
df.dropna(subset=["MOST_SEVERE_INJURY", 
                  "INJURIES_NON_INCAPACITATING", 
                  "INJURIES_UNKNOWN", 
                  "INJURIES_REPORTED_NOT_EVIDENT", 
                  "INJURIES_INCAPACITATING", 
                  "INJURIES_FATAL",
                  "INJURIES_TOTAL", 
                  "INJURIES_NO_INDICATION" 
                 ], axis=0, inplace=True)

# reset index, because we droped a number of rows
df.reset_index(drop=True, inplace=True)

print("We just deleted rows contaning missing labels. What's the shape of our dataframe now?")
print('Number of features: {}'.format(df.shape[1]))
print('Number of examples: {}'.format(df.shape[0]))
```

```python
# Create a function that converts all values of df['xxx'] into numbers

def weather_to_numeric(x):
    if x=='CLEAR':
        return 1
    if x=='CLOUDY/OVERCAST':
        return 2
    if x=='UNKNOWN':
        return 3
    if x=='OTHER':
        return 4
    if x=='BLOWING SAND, SOIL, DIRT':
        return 5
    if x=='RAIN':
        return 6
    if x=='SNOW':
        return 7
    if x=='SLEET/HAIL':
        return 8
    if x=='FREEZING RAIN/DRIZZLE':
        return 9
    if x=='BLOWING SNOW':
        return 10
    if x=='SEVERE CROSS WIND GATE':
        return 11
    if x=='FOG/SMOKE/HAZE':
        return 12
#####
#####
#####
def lighting_to_numeric(x):
    if x=='DAYLIGHT':
        return 1
    if x=='UNKNOWN':
        return 2
    if x=='DAWN':
        return 3
    if x=='DUSK':
        return 4
    if x=='DARKNESS, LIGHTED ROAD':
        return 5
    if x=='DARKNESS':
        return 6
#####
#####
#####
def first_crash_type_to_numeric(x):
    if x=='ANIMAL':
        return 1
    if x=='OTHER NONCOLLISION':
        return 2
    if x=='REAR END':
        return 3
    if x=='REAR TO REAR':
        return 4
    if x=='REAR TO FRONT':
        return 5
    if x=='REAR TO SIDE':
        return 6
    if x=='OTHER OBJECT':
        return 7
    if x=='FIXED OBJECT':
        return 8
    if x=='ANGLE':
        return 9
    if x=='TURNING':
        return 10
    if x=='SIDESWIPE OPPOSITE DIRECTION':
        return 11
    if x=='SIDESWIPE SAME DIRECTION':
        return 12
    if x=='PARKED MOTOR VEHICLE':
        return 13
    if x=='OVERTURNED':
        return 14
    if x=='TRAIN':
        return 15
    if x=='HEAD ON':
        return 16
    if x=='PEDALCYCLIST':
        return 17
    if x=='PEDESTRIAN':
        return 18
#####
#####
#####
def alignment_to_numeric(x):
    if x=='STRAIGHT AND LEVEL':
        return 1
    if x=='STRAIGHT ON GRADE':
        return 2
    if x=='CURVE, LEVEL':
        return 3
    if x=='STRAIGHT ON HILLCREST':
        return 4
    if x=='CURVE ON GRADE':
        return 5
    if x=='CURVE ON HILLCREST':
        return 6
#####
#####
#####
def roadway_surface_cond_to_numeric(x):
    if x=='DRY':
        return 1
    if x=='UNKNOWN':
        return 2
    if x=='OTHER':
        return 3
    if x=='SAND, MUD, DIRT':
        return 4
    if x=='WET':
        return 5
    if x=='SNOW OR SLUSH':
        return 6
    if x=='ICE':
        return 7
#####
#####
#####
def road_defect_to_numeric(x):
    if x=='NO DEFECTS':
        return 1
    if x=='UNKNOWN':
        return 2
    if x=='OTHER':
        return 3
    if x=='SHOULDER DEFECT':
        return 4
    if x=='WORN SURFACE':
        return 5
    if x=='RUT, HOLES':
        return 6
    if x=='DEBRIS ON ROADWAY':
        return 7
#####
#####
#####
def prim_contributory_cause_to_numeric(x):
    if x=='NOT APPLICABLE':
        return 1
    if x=='UNABLE TO DETERMINE':
        return 2
    if x=='ANIMAL':
        return 3
    if x=='EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST':
        return 4
    if x=='IMPROPER BACKING':
        return 5
    if x=='OBSTRUCTED CROSSWALKS':
        return 6
    if x=='BICYCLE ADVANCING LEGALLY ON RED LIGHT':
        return 7
    
    if x=='MOTORCYCLE ADVANCING LEGALLY ON RED LIGHT':
        return 8
    if x=='PASSING STOPPED SCHOOL BUS':
        return 9
    if x=='RELATED TO BUS STOP':
        return 10
    if x=='ROAD CONSTRUCTION/MAINTENANCE':
        return 11
    if x=='ROAD ENGINEERING/SURFACE/MARKING DEFECTS':
        return 12
    if x=='DISREGARDING YIELD SIGN':
        return 13
    if x=='IMPROPER TURNING/NO SIGNAL':
        return 14

    if x=='DRIVING SKILLS/KNOWLEDGE/EXPERIENCE':
        return 15
    if x=='EXCEEDING AUTHORIZED SPEED LIMIT':
        return 16
    if x=='EXCEEDING SAFE SPEED FOR CONDITIONS':
        return 17
    if x=='WEATHER':
        return 18
    if x=='VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)':
        return 19
    if x=='DISTRACTION - FROM OUTSIDE VEHICLE':
        return 20
    if x=='DISTRACTION - FROM INSIDE VEHICLE':
        return 21

    if x=='DISREGARDING TRAFFIC SIGNALS':
        return 22
    if x=='DISREGARDING STOP SIGN':
        return 23
    if x=='DISREGARDING ROAD MARKINGS':
        return 24
    if x=='DISREGARDING OTHER TRAFFIC SIGNS':
        return 25
    if x=='IMPROPER LANE USAGE':
        return 26
    if x=='FAILING TO REDUCE SPEED TO AVOID CRASH':
        return 27
    if x=='IMPROPER OVERTAKING/PASSING':
        return 28

    if x=='FOLLOWING TOO CLOSELY':
        return 29
    if x=='FAILING TO YIELD RIGHT-OF-WAY':
        return 30
    if x=='DRIVING ON WRONG SIDE/WRONG WAY':
        return 31
    if x=='DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)':
        return 32
    if x=='TURNING RIGHT ON RED':
        return 33
    if x=='CELL PHONE USE OTHER THAN TEXTING':
        return 34
    if x=='TEXTING':
        return 35

    if x=='HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)':
        return 36
    if x=='OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER':
        return 37
    if x=='EQUIPMENT - VEHICLE CONDITION':
        return 38
    if x=='PHYSICAL CONDITION OF DRIVER':
        return 39
    if x=='UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)':
        return 40
#####
#####
#####
# Apply our functions to the categorical variables 
df['WEATHER_CONDITION_TO_NUM'] = df['WEATHER_CONDITION'].apply(weather_to_numeric)
df['LIGHTING_CONDITION_TO_NUM'] = df['LIGHTING_CONDITION'].apply(lighting_to_numeric)
df['FIRST_CRASH_TYPE_TO_NUM'] = df['FIRST_CRASH_TYPE'].apply(first_crash_type_to_numeric)
df['ALIGNMENT_TO_NUM'] = df['ALIGNMENT'].apply(alignment_to_numeric)
df['ROADWAY_SURFACE_COND_TO_NUM'] = df['ROADWAY_SURFACE_COND'].apply(roadway_surface_cond_to_numeric)
df['ROAD_DEFECT_TO_NUM'] = df['ROAD_DEFECT'].apply(road_defect_to_numeric)
df['PRIM_CONTRIBUTORY_CAUSE_TO_NUM'] = df['PRIM_CONTRIBUTORY_CAUSE'].apply(prim_contributory_cause_to_numeric)

df
```

```python
# Create instance (i.e. object) of LogisticRegression
LogReg_model = LogisticRegression(C=0.01, multi_class='multinomial', solver='newton-cg', random_state=0)
print(LogReg_model)
print('')
print("----- OUTPUT BREAK -----")
print('')
print("Accuracy evaluation for logistic regression model")
# Step 1. Training . . . Fit the model using the training data
# X_train -> parameter supplies the data features
# y_train -> parameter supplies the target labels
LogReg_model.fit(X_train_transformed, y_train)
print("Train set, re: Accuracy classification score: %.6f" % metrics.accuracy_score(y_train, LogReg_model.predict(X_train_transformed)))
# 'Accuracy from sk-learn: {0}' returns the same value as metrics.accuracy_score
#print('Accuracy from sk-learn: {0}'.format(LogReg_model.score(X_train_transformed, y_train))) # print the training scores
# Step 2. Predicting 
yhat = LogReg_model.predict(X_test_transformed)
# Step 3. Accuracy evaluation
print("Jaccard similarity coefficient score: %.6f" % jaccard_score(y_test, yhat, average='weighted'))
print("F1 score: %.6f" % f1_score(y_test, yhat, average='weighted'))
yhat_prob = LogReg_model.predict_proba(X_test_transformed)
print("Log loss: %.6f" % log_loss(y_test, yhat_prob))
print('')
print("----- OUTPUT BREAK -----")
print('')
# Use scikit-learn classification_report to build a text report showing the main classification metrics
print("Classification report\n")
print(metrics.classification_report(y_test, yhat, digits=6))
print('')
print("----- OUTPUT BREAK -----")
print('')
# Quick check of predicted values
print('Quick check of predicted output from logistic regression model')
print(yhat[0:15])
```

Generalization of k-NN methodology . . . 

```python
# Find the best k 
Ks = 7
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    # Train the model and make predictions
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train_transformed,y_train)
    yhat=neigh.predict(X_test_transformed)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

Plot  model accuracy  for Different number of Neighbors 

```python
plt.plot(range(1,Ks),mean_acc,'blue')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (k)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

U.S. Firefighter Injuries in 2017. NFPA Journal.](https://www.nfpa.org/News-and-Research/Publications-and-Media/NFPA-Journal/2018/November-December-2018/Features/US-Firefighter-Injuries-2017). Some police traffic accidents are a direct result of first responders disregarding their department's policies or disregarding a direct order to terminate a pursuit. In Chicago, police pursuits have increased in frequency and resulted in more damage.

Chicago Sun-Times.)](https://chicago.suntimes.com/city-hall/2020/6/4/21280780/chicago-police-department-pursuit-policy-fatal-crash), Chicago's mayor as made swift changes to the police pursuit policy; this [policy was changed in August 2020](http://directives.chicagopolice.org/directives/data/a7a57be2-1291920c-54712-9192-aecb02ef17e0c47d.pdf?hl=true)

First responders involved in traffic accidents en route to incident calls have immeasurable costs to those requesting help along with those answering the call for help. It is in everyone's interest to know and understand the facts of this situation. With a shared understanding of the facts, the general public, elected officials and first responders can develop data-driven policies to mitigate accidents during first responders' incident calls.

This analysis will use **supervised machine learning (SML)** to predict the severity of accidents. In addition, through non-SML statistical modeling I will attempt to determine if first responders / all other vehicles have a statistically significant impact on accident severity.
